# Question
I collected last 3 years of students scores and study hour data. <br>
I want to know how many hours do I have to study to get 80 <br>

# Variables <br>
<li>m = number of points</li>
<li>X[i] = # hour studied by ith student </li>
<li>Y[i] = math score of ith student</li>
<li>W = Parameters</li>
<li>dw = partial derivative of w</li>
<li>db = partial derivative of b</li>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
'''
Assume that the data are generated by the

y = W[0] * x + W[1]

We want to find W
'''

data = np.array(pd.read_csv("data/1.csv", header=None))

X = np.array(data[:, 0])
Y = np.array(data[:, 1])
m = len(X)

plt.xlabel("hours studied")
plt.ylabel("score")
plt.scatter(X, Y)
```

```python
def y_hat(x, W):
    return W[0] + W[1] * x

# mean square error
def compute_cost(W):
    cost = 0
    
    for i in range(len(X)):
        cost += (y_hat(X[i], W) - Y[i]) ** 2
    
    return cost / 2 / m

def partial_deriv(W):
    dW = np.zeros(2)
    
    for i in range(m):
        dW[0] += (y_hat(X[i], W) - Y[i])
        dW[1] += (y_hat(X[i], W) - Y[i]) * X[i]
    
    dW /= m
    
    return dW

def model(learn_rate, n_iters, W):
    Cost = []
    
    for i in range(n_iters):
        W -= learn_rate * partial_deriv(W)
        Cost.append(compute_cost(W))
    
    return W, Cost

def draw(W):
    plt.scatter(X, Y)
    
    X_temp = np.linspace(0, 10, 1000)
    plt.scatter(X_temp, [y_hat(x_temp, W) for x_temp in X_temp], s=1)
    plt.xlabel("hours studied")
    plt.ylabel("score")
    plt.show()

W, Cost = model(0.05, 1000, [0, 0])  # play with learn_rate, n_iters
print("score = {:.1f} + {:.1f} X".format(W[0], W[1]))
draw(W)

plt.xlabel("iteration")
plt.ylabel("cost")
plt.plot(Cost)
```

So, in order to get 80 <br>
I need to study about 8 hours (33.3 + 5.9 * 8) = 80.5
