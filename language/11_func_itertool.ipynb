```python
def flatten(obj):
    if isinstance(obj, str) or isinstance(obj, int):
        return str(obj)
    elif isinstance(obj, list):
        return " ".join(flatten(e) for e in obj)
    elif isinstance(obj, dict):
        return " ".join(flatten(e) for e in obj.values())

def grouper(n, iterable, fillvalue=None):
    args = [iter(iterable)] * n
    return list(itertools.zip_longest(fillvalue=fillvalue, *args))
```

```python
print(flatten([1, 2, 3, [1, 2, 3]]))
print(grouper(3, [1,2,3,4,5,6]))
```

```python
import functools
```

```python
@functools.lru_cache(maxsize=128)
def cache_fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return cache_fib(n - 1) + cache_fib(n - 2)

def nocache_fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return nocache_fib(n - 1) + nocache_fib(n - 2)
```

```python
start = time.perf_counter()
for _ in range(100):
    cache_fib(22)

pprint(f'Took {time.perf_counter() - start}')
start = time.perf_counter()

for _ in range(100):
    nocache_fib(22)

pprint(f'Took {time.perf_counter() - start}')
```

```python
def make_blink(function):
    @functools.wraps(function)
    def decorator():
        return f"<p>{function()}</p>"
    return decorator

pprint(f"greet() \t {greet()}")
pprint(f"greet.__name__ \t {greet.__name__}")
pprint(f"greet.__doc__ \t {greet.__doc__}")

@make_blink
def greet():
    """prints Hello"""
    return "Hello"

pprint("\nwith @make_blink")
pprint(f"greet() with \t {greet()}")
pprint(f"greet.__name__ \t {greet.__name__}")
pprint(f"greet.__doc__ \t {greet.__doc__}")
```
