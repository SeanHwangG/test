```python
import sysconfig
```

```python
def show_info():
    conf = sysconfig.get_config_vars()
    print(conf['BLDSHARED'])
    print("CFLAGS")
    print(sysconfig.get_config_var('CFLAGS'))
    print(sysconfig.get_config_var('LIBDIR'))
    print(sysconfig.get_config_vars('AR', 'CXX'))
show_info()
```
