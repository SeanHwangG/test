```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
pprint = lambda st : print(st.expandtabs(32))
```

```python
corpus = [
     'AA BB CC',
     'CC DD',
]
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
display(df)
```

```python
corpus = [
     'AA BB AA',
]
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
display(df)
```

```python
corpus = [
     'This', 'is', 'the', 'first', 'the'
]
vectorizer = TfidfVectorizer()
vectorizer.fit(corpus)
vectors = vectorizer.transform(corpus)

feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
display(df)
```

## Label Encoder

```python
le = preprocessing.LabelEncoder()
le.fit([1, 2, 2, 6])
pprint(f"All classes \t {le.classes_}")
pprint(f"transform \t {le.transform([6, 1, 1])}")
pprint(f"Inverse_transform \t {le.inverse_transform([0, 1])}")
```

## metrics

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import pairwise_distances
```

```python
y_pred = [0, 2, 1, 3]
y_true = [0, 1, 2, 3]
pprint(f"jaccard_similarity \t {jaccard_similarity_score(y_true, y_pred)}")
pprint(f"jaccard_similarity norm \t {jaccard_similarity_score(y_true, y_pred, normalize=False)}")

X = np.array([[1, 0],  
              [1, 2]])

Y = np.array([[2, 0], 
              [2, 1]])

pprint(f"pairwise_distances \t {pairwise_distances(X, Y, metric='manhattan').tolist()}")
pprint(f"pairwise_distances \t {pairwise_distances(X, Y, metric='cosine').tolist()}")
pprint(f"pairwise_distances \t {pairwise_distances(X, Y, metric='jaccard').tolist()}")
```
